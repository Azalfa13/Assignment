```python
print("Calculator")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

operation = input("Enter choice(1/2/3/4): ")

if operation == '1':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

if operation == '2':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")

if operation == '3':
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")

if operation == '4':
    if num2 == 0:
        print("Error! Cannot divide by zero.")
    else:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
```

```python
my_list = ["apple", 10, "banana", 3.14, "cherry", True]

found_numeric = False

print(f"Checking the list: {my_list}")

for item in my_list:
    if isinstance(item, (int, float)):
        print(f"Found a numeric value: {item}")
        found_numeric = True 

if not found_numeric:
    print("No numeric values were found in the list")
```

```python
my_dictionary = {"name": "Alice", "age": 30}

print("Original dictionary:", my_dictionary)

my_dictionary["city"] = "New York"

print("Dictionary after adding 'city':", my_dictionary)

my_dictionary["occupation"] = "Engineer"

print("Dictionary after adding 'occupation':", my_dictionary)
```

```python
my_dict = {
    'item1': 10,
    'item2': 'hello',
    'item3': 20.5,
    'item4': True,  
    'item5': 'world',
    'item6': -5
}

total_sum = 0

print(f"Original dictionary: {my_dict}")

for value in my_dict.values():
    if isinstance(value, (int, float)):
        total_sum += value

print(f"Sum of numeric items: {total_sum}")
```

```python
my_list = [1, 2, 3, 2, 1, 4, 5, 3, 6, 7, 7, 8]

seen_elements = set()
duplicate_elements = []

print(f"Original list: {my_list}")

for item in my_list:
    if item in seen_elements:
        if item not in duplicate_elements:
            duplicate_elements.append(item)
    else:
        seen_elements.add(item)

if duplicate_elements:
    print(f"Duplicate values found: {duplicate_elements}")
else:
    print("No duplicate values found in the list.")
```

```python
my_dict = {"name": "Alice", "age": 30, "city": "New York"}

key_to_check = "age"
key_not_in_dict = "country"

print(f"Checking dictionary: {my_dict}")

if key_to_check in my_dict:
    print(f"The key '{key_to_check}' exists in the dictionary. Its value is: {my_dict[key_to_check]}")
else:
    print(f"The key '{key_to_check}' does not exist in the dictionary.")

if key_not_in_dict in my_dict:
    print(f"The key '{key_not_in_dict}' exists in the dictionary. Its value is: {my_dict[key_not_in_dict]}")
else:
    print(f"The key '{key_not_in_dict}' does not exist in the dictionary.")
```
