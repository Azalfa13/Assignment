```python
subject_marks = []

print("Enter marks for 5 subjects:")
for i in range(1, 6):
    while True:
        try:
            mark = float(input(f"Enter marks for Subject {i}: "))
            if 0 <= mark <= 100:
                subject_marks.append(mark)
                break
            else:
                print("Marks must be between 0 and 100. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a numerical value for marks.")

print("Subject marks entered:", subject_marks)
total_marks = sum(subject_marks)
percentage = (total_marks / 500) * 100

if percentage >= 90:
    grade = 'A+'
elif percentage >= 80:
    grade = 'A'
elif percentage >= 70:
    grade = 'B'
elif percentage >= 60:
    grade = 'C'
elif percentage >= 50:
    grade = 'D'
else:
    grade = 'F'

print("\n--- Marksheet ---")
for i, mark in enumerate(subject_marks):
    print(f"Subject {i+1}: {mark:.2f}/100")
print(f"Total Marks: {total_marks:.2f}/500")
print(f"Percentage: {percentage:.2f}%")
print(f"Grade: {grade}")
```
